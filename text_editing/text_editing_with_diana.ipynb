```python
def read_text_file(text):
    # Remove chapter title and other information
    
    #"substitutes everything non alphanumeric into nothing, and things between parenthesis for example go away"
    text = re.sub(r'\b[A-Z]+\s+\d+|\([\w\s\d:;-]+\)', '', text)
    #removing new line
    text = re.sub(r'\n', '  ', text)
    #splitting based on period, ! or ?
    sentences = re.split(r'(?<=[.!?])\s+', text)
    #removing numbers
    sentences = [re.sub(r'\d+', '', s) for s in sentences if s]

    substring = "    "
    output = []
    for s in sentences:
        if substring in s:
            output.append(s[s.index(substring)+len(substring):])
        else: 
            output.append(s)
    
    
    for i, sentence in enumerate(output):
        if re.match(r'^\W', sentence):
            output[i] = re.sub(r'^\W+', '', sentence)

    return output

```

```python
def bible_process_pack(name, text):
    bible_text_list = read_text_file(text)
    bible_df = pd.DataFrame(bible_text_list, columns = ['first'])
    bible_df['second'] = bible_df['first'].shift(-1)
    bible_df['label'] = 1
    bible_df = bible_df[:][:-1]
    file_name_correct = name + "_correct" + ".csv"
    bible_df.to_csv(file_name_correct)

    incorrect = []
    txt_len = len(bible_df)
    for idx in range(len(bible_df)):
        second_idx = random.randint(0, txt_len - 1)
        if second_idx == idx: 
        #avoid number that is equal to the first sentence
            second_idx += 1
            
        if second_idx == idx + 1:
        #or the one after it
            second_idx += 1
            
        incorrect.append(bible_text_list[second_idx])
    bdf_incorrect = bible_df.copy()
    bdf_incorrect['second'] = incorrect
    bdf_incorrect['label'] = 0

    file_name_incorrect = name + "_incorrect" + ".csv"
    bdf_incorrect.to_csv(file_name_incorrect)
    

```

```python
for i in bibles.keys():
    bible_process_pack(i, bibles[i])
```
